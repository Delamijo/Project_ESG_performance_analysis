# Correlations

## Imports

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats
from pathlib import Path
```

## Data

```python
base_ordner = Path().resolve().parent
proc_data_ordner = base_ordner / "data" / "processed"

df_l = pd.read_csv(proc_data_ordner / "lauschke_kpis.csv")
df_bench = pd.read_csv(proc_data_ordner / "benchmark.csv")
```

## Correlation

```python
corr_margin, p_margin = stats.pearsonr(df_l["ESG_Overall"], df_l["ProfitMargin"])
corr_market, p_market = stats.pearsonr(df_l["ESG_Overall"], df_l["MarketCap"])
corr_co2, p_co2 = stats.pearsonr(df_l["CO2_intensity"], df_l["ProfitMargin"])

print(f"ESG Overall vs Profit Margin: r={corr_margin:.3f}, p={p_margin:.3f}")
print(f"ESG Overall vs Market Cap: r={corr_market:.3f}, p={p_market:.3f}")
print(f"CO2 Intensity vs Profit Margin: r={corr_co2:.3f}, p={p_co2:.3f}")
```

### Q3

ESG Overall vs Profit Margin: Strong positive correlation, statistically significant. Higher ESG scores are associated with higher profit margins over the 2015-2025 period.

ESG Overall vs MarketCap: No meaningful correlation, not significant.

CO2 Intensity vs Profit Margin: Strong negative correlation, significant. Lower CO2 intensity is associated with higher profit margins.

Of course correlation does not imply causation and was not corrected for confounding variables. n=11 limits the statistical power.

## Benchmark for european Market

```python
# Querschnitt 2025
df_25 = df_bench[df_bench["Year"] == 2025]
corr_margin_b, p_margin_b = stats.pearsonr(df_25["ESG_Overall"], df_25["ProfitMargin"])
corr_market_b, p_market_b = stats.pearsonr(df_25["ESG_Overall"], df_25["MarketCap"])

df_25["CO2_intensity"] = (df_25["CarbonEmissions"] / df_25["Revenue"]).round(2)
corr_co2_b, p_co2_b = stats.pearsonr(df_25["CO2_intensity"], df_25["ProfitMargin"])

print(f"ESG Overall vs Profit Margin: r={corr_margin_b:.3f}, p={p_margin_b:.3f}")
print(f"ESG Overall vs Market Cap: r={corr_market_b:.3f}, p={p_market_b:.3f}")
print(f"CO2 Intensity vs Profit Margin: r={corr_co2_b:.3f}, p={p_co2_b:.3f}")
```

```python
# per company
results = []
for idx, grp in df_bench.groupby("CompanyID"):
    r, p = stats.pearsonr(grp["ESG_Overall"], grp["ProfitMargin"])
    results.append({"CompanyID": idx, "r": r, "p": p})

df_corr = pd.DataFrame(results)
print(f"Median r: {df_corr['r'].median():.3f}")
```

```python
df_corr.to_csv(proc_data_ordner / "corr_per_company.csv")
df_25.to_csv(proc_data_ordner / "bench_25.csv")

print("done")
```

The strong correlation in Lauschke GmbHs case seems to be company specific and does not seem to be generalized across european maufactuerers.
